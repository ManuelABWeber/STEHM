```python
from ultralytics import YOLO
model = YOLO('yolov8l.pt')

results = model.train(data='/content/drive/MyDrive/Yolo8/data.yaml',
                      epochs=50,
                      patience=15,
                      imgsz=640,
                      batch=16,
                      name="yolo81704",
                      plots=True,
                      save_period=10,  # Save a checkpoint every 5 epochs.
                      fliplr=0.5,  # Horizontal flip probability
                      scale=0.3,  # Random scaling
                      perspective=0.0003,  # Slight perspective change
                      translate=0.1,  # Random translation
                      hsv_h=0.015,  # Hue shift
                      hsv_s=0.7,  # Saturation fluctuation
                      hsv_v=0.4,  # Value fluctuation
                      lr0=0.004,  # Initial learning rate.
                      lrf=0.1,  # Final learning rate as a factor of the initial rate.
                      momentum=0.937,  # Momentum.
                      weight_decay=0.0005,  # Weight decay for regularization.
                  )
```
